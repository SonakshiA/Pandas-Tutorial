```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
filt = (df['last']=='Doe') #returns a series object
```

```python
filt
```

```python
df[filt]
```

```python
df[df['last']=='Doe'] #another way to do it
```

```python
df.loc[filt] #another way to do it, Passing a series of boolean to filter data out
```

```python
df.loc[filt,'email'] #Series object
```

```python
# we can't use python built-in and or for filtering. We use & |
filt = (df['last']=='Doe') | (df['last'] == 'Schafer')
df[filt]
```

```python
filt = (df['last']=='Doe') & (df['first'] == 'Jane')
df[filt]
```

```python
df.loc[~filt,'email'] #get the opposite of the filter (everything that did not match the filter)
```
