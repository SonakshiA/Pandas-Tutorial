```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.sort_values(by='last') #sort in ascending order
```

```python
df.sort_values(by='last', ascending=False) #sort in descending order
```

```python
df.sort_values(by=['first','last'], ascending=False)
```

```python
df.sort_values(by=['first','last'], ascending=[False,True]) 
#first column sorted by descending and last column sorted by ascending
```

```python
df.sort_index()
```

```python
df['last'].sort_values() #sorting on a Series object
```
